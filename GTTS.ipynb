```python
from paddleocr import PaddleOCR, draw_ocr
import cv2
from gtts import gTTS
from googletrans import Translator
import numpy as np
from IPython.display import Audio, display

# Initialize PaddleOCR with English model
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Initialize the Translator
translator = Translator()

# Function to perform OCR on a given image
def ocr_on_image(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Perform OCR on the image
    result = ocr.ocr(image, cls=True)

    # Extract text
    texts = [line[1][0] for line in result[0]]
    extracted_text = " ".join(texts)

    return extracted_text, image, result

# Function to translate text to a specified language
def translate_text(text, target_lang):
    translated = translator.translate(text, dest=target_lang)
    return translated.text

# Function to convert text to speech
def text_to_speech(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save("output.mp3")
    return "output.mp3"

# Main function to combine everything
def main(image_path, target_lang):
    # Perform OCR on the provided image
    text, image, ocr_result = ocr_on_image(image_path)

    # Display the extracted text
    print("Extracted Text:")
    print(text)

    # Translate the extracted text to the target language
    translated_text = translate_text(text, target_lang)
    print(f"Translated Text ({target_lang}):")
    print(translated_text)

    # Convert the translated text to speech and play it
    audio_file = text_to_speech(translated_text, target_lang)
    display(Audio(audio_file, autoplay=True))

    # Optional: Display the image with detected text boxes
    boxes = [line[0] for line in ocr_result[0]]
    annotated_image = draw_ocr(image, boxes, [line[1][0] for line in ocr_result[0]], [line[1][1] for line in ocr_result[0]])
    annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_RGB2BGR)
    cv2.imshow('OCR Result', annotated_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# ... (rest of the code)

if __name__ == "__main__":
    # Provide the path to your image here
    image_path = '/content/images.jpeg'
    # Specify the target language ('mr' for Marathi, 'ta' for Tamil, 'ml' for Malayalam)
    target_languages = ['mr', 'hi', 'ta']    # Example: 'mr' for Marathi, 'ta' for Tamil, 'ml' for Malayalam
    selected_language = input("Choose the language to visualize (mr for Marathi, hi for Hindi, ta for Tamil): ")
    if selected_language not in target_languages:
        print("Invalid language choice. Please run the script again and select a valid language.")
    else:
        # Call main with the selected language as a string, not the entire list
        main(image_path, selected_language)
```

```python
import streamlit as st
from paddleocr import PaddleOCR, draw_ocr
from googletrans import Translator
import matplotlib.pyplot as plt
from PIL import Image
import os

# Initialize the OCR model
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Initialize the translator
translator = Translator()

# Function to perform OCR on the image
def ocr_on_image(image):
    result = ocr.ocr(image, cls=True)
    detected_texts = [elements[1][0] for line in result for elements in line]
    boxes = [elements[0] for line in result for elements in line]
    scores = [elements[1][1] for line in result for elements in line]
    return detected_texts, boxes, scores

# Function to translate text into chosen languages
def translate_texts(detected_texts, target_languages):
    translations = {lang: [translator.translate(text, dest=lang).text for text in detected_texts] for lang in target_languages}
    return translations

# Streamlit UI
st.title("Real-Time Text Detection and Translation")
st.write("Upload an image, select a language, and see the translated text on the image.")

# File upload
uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file).convert('RGB')

    # Perform OCR on the image
    detected_texts, boxes, scores = ocr_on_image(uploaded_file)

    # Define target languages for translation
    target_languages = {
        'Marathi': 'mr',
        'Hindi': 'hi',
        'Tamil': 'ta'
    }

    # Select target language
    selected_language = st.selectbox("Choose the language to visualize:", list(target_languages.keys()))
    target_lang_code = target_languages[selected_language]

    # Translate the detected text into the selected language
    translations = translate_texts(detected_texts, [target_lang_code])
    translated_texts = translations[target_lang_code]

    # Draw the results on the image
    font_path = os.path.join('PaddleOCR', 'doc', 'fonts', 'latin.ttf')
    im_show = draw_ocr(np.array(image), boxes, translated_texts, scores, font_path=font_path)
    im_show = Image.fromarray(im_show)

    # Display the image with translated text
    st.image(im_show, caption='Translated Image', use_column_width=True)

    # Print the translations
    st.write("Translations:")
    for original, translated in zip(detected_texts, translated_texts):
        st.write(f'**Original:** {original} -> **Translated:** {translated}')
```
